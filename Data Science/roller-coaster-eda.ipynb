# Step 4: Feature Relationships
- Scatterplot
- Heatmap Correlation
- Pairplot

```python
df.plot(kind = "scatter", 
       x = "speed_mph",
       y = "height_ft",
       title = "Coaster speed vs height")
plt.show()
```

```python
sns.scatterplot(x = "speed_mph", y = "height_ft", hue = "year_introduced", data=  df)
```

```python
sns.pairplot(df, 
             vars = ["year_introduced", "speed_mph", "height_ft", "Inversions_clean","Gforce_clean"],
             hue = "Type_Main")
```

```python
df_corr = df[["year_introduced", "speed_mph", "height_ft", "Inversions_clean", "Gforce_clean"]].dropna().corr()
```

```python
sns.heatmap(df_corr, annot = True)
```

# Step 5: Ask a Question about the data

What are the locations with the fastest roller coasters (minimum of 10)?

```python
df.loc[df.speed_mph >= 10, "year_introduced"].unique()
```

```python
df.loc[df.Location != "Other", :].groupby()["speed_mph"].agg(["mean", "count"]).query("count >= 10").sort_values('mean').plot(kind = "barh")
```
